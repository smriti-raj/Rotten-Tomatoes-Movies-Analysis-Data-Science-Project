```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv(r"S:\Data Set\Excel dataset\Excel dataset\Rotten Tomatoes Movies.csv")

df.head()
```

```python
print("Rows :", df.shape[0])
print("Columns :", df.shape[1])
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.describe(include='all')
```

```python
df.columns.tolist()
```

```python
#Top 10 Highest Rated Movies 

top_movies = df.sort_values(
    by='tomatometer_rating',
    ascending=False
).head(10)

top_movies[['movie_title','tomatometer_rating']]
```

```python
# Top Rated Movies Chart
plt.figure(figsize=(10,6))

sns.barplot(
    x='tomatometer_rating',
    y='movie_title',
    data=top_movies
)

plt.title("Top Rated Movies")
plt.show()
```

```python
# Tomatometer Rating Distribution
plt.figure(figsize=(8,5))

sns.histplot(
    df['tomatometer_rating'],
    bins=20,
    kde=True
)

plt.title("Tomatometer Rating Distribution")
plt.show()
```

```python
## Tomatometer Status
status = df["tomatometer_status"].value_counts()

plt.figure(figsize=(8,5))

plt.bar(
    status.index,
    status.values,
    color="limegreen",
    edgecolor="black"
)

plt.title("Tomatometer Status")

plt.show()
```

```python
# Pie Chart of Tomatometer Status
status = df["tomatometer_status"].value_counts()

plt.figure(figsize=(8,8))

plt.pie(
    status,
    labels=status.index,
    autopct="%1.1f%%",
    startangle=90,
    shadow=True
)

plt.title("Tomatometer Status")

plt.show()
```

```python
# Audience Rating Distribution
plt.figure(figsize=(8,5))

sns.histplot(
    df['audience_rating'],
    bins=20,
    kde=True
)

plt.title("Audience Rating Distribution")
plt.show()
```

```python
## Audience Count Distribution
plt.figure(figsize=(10,5))

plt.hist(
    df["audience_count"].dropna(),
    bins=30,
    color="teal",
    edgecolor="black"
)

plt.title("Audience Count Distribution")

plt.show()
```

```python
# Audience Rating vs Tomatometer Rating
plt.figure(figsize=(8,6))

plt.scatter(
    df["audience_rating"],
    df["tomatometer_rating"],
    color="deeppink",
    alpha=0.5
)

plt.xlabel("Audience Rating")

plt.ylabel("Tomatometer Rating")

plt.title("Audience Rating vs Tomatometer Rating")

plt.grid(True)

plt.show()
```

```python
# Runtime vs Audience Rating
plt.figure(figsize=(8,6))

plt.scatter(
    df["runtime_in_minutes"],
    df["audience_rating"],
    color="darkgreen",
    alpha=0.5
)

plt.xlabel("Runtime (Minutes)")

plt.ylabel("Audience Rating")

plt.title("Runtime vs Audience Rating")

plt.show()
```

```python
# Runtime vs Tomatometer Rating
plt.figure(figsize=(8,6))

plt.scatter(
    df["runtime_in_minutes"],
    df["tomatometer_rating"],
    color="royalblue",
    alpha=0.5
)

plt.xlabel("Runtime")

plt.ylabel("Tomatometer Rating")

plt.title("Runtime vs Tomatometer Rating")

plt.show()
```

```python
# Top Rated Movies Chart
plt.figure(figsize=(10,6))

sns.barplot(
    x='tomatometer_rating',
    y='movie_title',
    data=top_movies
)

plt.title("Top Rated Movies")
plt.show()
```

```python
# Runtime Distribution
plt.figure(figsize=(10,5))

plt.hist(
    df["runtime_in_minutes"].dropna(),
    bins=25,
    color="purple",
    edgecolor="black"
)

plt.title("Movie Runtime Distribution")

plt.show()
```

```python
# Audience vs Critics Score
import seaborn as sns
plt.figure(figsize=(8,6))

plt.scatter(
    df["audience_rating"],
    df["tomatometer_rating"],
    color="royalblue",
    alpha=0.5
)

plt.xlabel("Audience Rating")

plt.ylabel("Tomatometer Rating")

plt.show()
```

```python
#Average Rating by Genre
genre_avg = df.groupby(
    'genre'
)['tomatometer_rating'].mean().sort_values(
    ascending=False
)

genre_avg.head(10)
```

```python
# Genre Rating Chart
plt.figure(figsize=(10,6))

genre_avg.head(10).plot(kind='bar')

plt.title("Best Rated Genres")
plt.ylabel("Average Rating")
plt.show()

```

```python
# Directors with Most Movies
top10 = df["directors"].value_counts().head(10)

top10.plot(
    kind="bar",
    figsize=(12,5),
    color="darkorange"
)

plt.title("Top Directors")

plt.show()
```

```python
# Writers
df["writers"].value_counts().head(10).plot(
    kind="barh",
    figsize=(10,6),
    color="purple"
)

plt.title("Top Writers")

plt.show()
```

```python
# Movies Released Per Year
release = df["in_theaters_date"].value_counts().sort_index()

plt.figure(figsize=(12,5))

plt.plot(
    release.index,
    release.values,
    color="crimson",
    linewidth=3,
    marker="o"
)

plt.title("Movie Release Trend", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Number of Movies")

plt.grid(True)

plt.show()
```

```python
# Release Trend
plt.figure(figsize=(12,5))

df['in_theaters_date'].value_counts().sort_index().plot()

plt.title("Movies Released Over Years")
plt.show()
```

```python
# Top Studios
top10 = df["studio_name"].value_counts().head(10)

plt.figure(figsize=(12,6))

plt.barh(
    top10.index,
    top10.values,
    color="green"
)

plt.title("Top Studios")

plt.show()
```

```python
# Studios Pie Chart
studio = df["studio_name"].value_counts().head(5)

plt.figure(figsize=(8,8))

plt.pie(
    studio,
    labels=studio.index,
    autopct="%1.1f%%"
)

plt.show()
```

```python
#Correlation Heatmap
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```
